```python
import os
import fitz
import pytesseract
from PIL import Image
import io

pytesseract.pytesseract.tesseract_cmd = r"R:\Manager( Program )\tesseract.exe"

doc = fitz.open("imp question.pdf")

text = ""

for page in doc:
    pix = page.get_pixmap()
    img_data = pix.tobytes("png")

    image = Image.open(io.BytesIO(img_data))

    text += pytesseract.image_to_string(image)

print(text)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(chunk_size = 100, chunk_overlap = 20)
chunk = splitter.create_documents([text])
```

```python
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings


embeddings = HuggingFaceEmbeddings()
vector_store = FAISS.from_documents(chunk, embeddings)
```

```python
retrival = vector_store.as_retriever( search_kwargs={"k": 3},  
    search_type="similarity"
)
```

```python
from langchain_core.prompts import PromptTemplate
```

```python
prompt = PromptTemplate(
    template="""
    One the pdf question
    {context}
    question:{question}
"""
)

question = 'Important question of the Data strucutres and algorithm for 10 marks'
retrivaldoc = retrival.invoke(question)
```

```python
from dotenv import load_dotenv
import os
```

```python
content_text = '\n\n'.join(doc.page_content for doc in retrivaldoc)
```

```python
from langchain_groq import ChatGroq

final_prompt = prompt.invoke({'context': content_text, 'question': question})

llm = ChatGroq(model='llama-3.1-8b-instant')
answer = llm.invoke(final_prompt)
print(answer.content)
```
